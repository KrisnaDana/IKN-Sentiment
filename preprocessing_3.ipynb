```python
# IMPORT LIBRARY
from nltk.tokenize import word_tokenize
from pymongo import MongoClient
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# IMPORT DATA
client = MongoClient("mongodb://localhost:27017/")
collection = client["data"]["preprocessing_2"]
data = collection.find()

# GET COMMENT AND DATE ONLY
data = collection.find()
data = [d for d in data]
printer = [print(str(d['date']) + " " + str(d['text'])) for d in data]
```

```python
# EXPORT DATA FOR TF-IDF TO MONGODB

# TOKENIZE TO WORDS
tf_idf_word = [d['text'] for d in data]
tf_idf_data = [[word, data[i]['date']] for i, word in enumerate(tf_idf_word)]

# SAVE TO MONGODB
tf_idf_collection = client["data"]["tf_idf"]
tf_idf_collection.insert_many({'comment': i[0], 'date':i[1]} for i in tf_idf_data)
```

```python
# EXPORT DATA LABELLING TO CSV AND EXPORT REMAINING DATA TO MONGODB
remaining_data, label_data = train_test_split(data,test_size=0.1)

# EXPORT LABEL DATA TO CSV
df_label = pd.DataFrame(label_data, columns=['_id', 'text', 'date'])
df_label = df_label.drop(['_id'], axis=1)
df_label.to_excel('data/not_yet_labelled_data.xlsx', index=False)

# EXPORT REMAINING DATA TO MONGODB FOR NAIVE BAYES
remaining_collection = client["data"]["remaining"]
df_remaining = pd.DataFrame(remaining_data, columns=['_id', 'text', 'date'])
df_remaining = df_remaining.drop(['_id'], axis=1)
remaining_data = df_remaining.to_dict(orient='records')
remaining_collection.insert_many(remaining_data)
```
