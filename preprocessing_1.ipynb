```python
# IMPORT LIBRARY
from pymongo import MongoClient
import pandas as pd
```

```python
# IMPORT DATA
df = pd.read_excel('data/data_kotor.xlsx', sheet_name=0)
display(df)
```

```python
# RERMOVE NAN VALUE
df['textOriginal'].dropna(inplace=True)
df['updatedAt'].dropna(inplace=True)
display(df)
```

```python
# COMBINE AND RENAME COLUMN
df.rename(columns = {'textOriginal':'text', 'updatedAt':'date'}, inplace = True)
df = df.drop(['videoId', 'textDisplay', 'authorDisplayName',
              'authorProfileImageUrl', 'authorChannelUrl', 'authorChannelId',
              'canRate', 'viewerRating', 'likeCount', 'publishedAt', 'parentId', 'commentId'
              ], axis=1)
display(df)
```

```python
# REMOVE DUPLICATE COMMENTS
df = df.drop_duplicates(subset=['text'])
display(df)
```

```python
# SORT BY DATE
df = df.sort_values('date')
display(df)
```

```python
# SAVE TO MONGODB
client = MongoClient("mongodb://localhost:27017/")
collection = client["data"]["preprocessing_1"]
data = df.to_dict(orient='records')
collection.insert_many(data)
```
